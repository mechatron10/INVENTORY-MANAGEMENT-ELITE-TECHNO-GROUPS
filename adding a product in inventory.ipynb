```python
#adding a new product in inventory
import json
fd=open("new displayinventory.json",'r') 
t=fd.read()
fd.close()
invent=json.loads(t)
```

```python
invent
```

```python
n=int(input("enter no. of products to be added:"))
li=[]
for i in invent:
    li.append(i)

for i in range (0,n):
    pid=input("enter product id:")
    b=input("enter product name:")
    na=input("enter brand name:")
    pr=int(input("enter the price of product:"))
    ex=input("enter the expiry date:")
    w=input("enter its weight:")
    q=int(input("enter no. of products:"))
for i in range(0,len(invent)):
    if(li[i]==pid):
                a=invent[str(pid)]['quantity']
                invent[str(pid)]['quantity']=a+q
                break
    else:
        invent[pid]={'name':b,'brand':na,'price':pr,'expiry':ex,'wieght':w,'quantity':q}
            
```

```python
print("products mentioned are added")        
b=json.dumps(invent)
fd=open("new displayinventory.json",'w')
fd.write(b)
fd.close()
```
